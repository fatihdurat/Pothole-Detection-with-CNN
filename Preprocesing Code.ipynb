```python
import numpy as np
import glob
import cv2
import os
```

```python
def PreProcessing(image, NewFolder, String, i):
    img = cv2.imread(image,cv2.IMREAD_COLOR)
    img = img[1200:1900,500:3000]
    img = cv2.resize(img, (200,200))
    cv2.imwrite(NewFolder +"/{}_{}.JPG".format(String,i), img)
    return img
```

```python
Images = []
Labels = []

path = "/content/drive/MyDrive/DATASET"
os.mkdir("/content/drive/MyDrive/Resized Data2")
DatasetFolders = os.listdir(path)
MainFolder = "/content/drive"
i= 1
for SubFolders in DatasetFolders:
    os.chdir(path)
    if SubFolders == "Negative data":
        os.mkdir("/content/drive/MyDrive/Resized Data2"+"/Negative Data")
        ResizedNegativeFolder = "/content/drive/MyDrive/Resized Data2"+"/Negative Data"
        os.chdir(SubFolders)
        for img in glob.glob(os.getcwd() + "/*.JPG"):
            Labels.append("Negative")
            Images.append(np.array(PreProcessing(img,ResizedNegativeFolder,"Negative",i)))
            i = i+1
        os.chdir(MainFolder)
    
    
    else:
        os.mkdir("/content/drive/MyDrive/Resized Data2"+"/Positive Data")
        ResizedPositiveFolder = "/content/drive/MyDrive/Resized Data2"+"/Positive Data"
        os.chdir(SubFolders)
        for img in glob.glob(os.getcwd() + "/*.JPG"):
            Labels.append("Positive")
            Images.append(np.array(PreProcessing(img,ResizedPositiveFolder,"Positive",i)))
            i = i+1
        os.chdir(MainFolder)
print(i)
```
